## Question 1: Variable Scope & Reference

**Problem:** Understand how variable scope works in Python

```python
# Question 1
x = 5
def modify():
    x = 10
    return x

print("Function returns:", modify())
print("Global x:", x)
print()
print("EXPLANATION:")
print("- Function creates LOCAL variable x = 10")
print("- Global x remains 5")
print("- Local scope doesn't affect global scope")
```

## Question 2: List Mutability & Reference

**Problem:** Understanding mutable vs immutable types

```python
# Question 2
list1 = [1, 2, 3]
list2 = list1  # Assignment: both reference same object
list2[0] = 99

print("list1:", list1)
print("list2:", list2)
print()
print("EXPLANATION:")
print(f"list1 is list2: {list1 is list2}")
print("- Assignment (=) creates reference, not copy")
print("- Both variables point to SAME object in memory")
print("- Modifying one affects the other")
print()
print("HOW TO FIX:")
list1 = [1, 2, 3]
list2 = list1.copy()  # or list(list1) or list1[:]
list2[0] = 99
print("After using .copy():")
print("list1:", list1)
print("list2:", list2)
```

## Question 3: Dictionary Key Types

**Problem:** Only hashable types can be dict keys

```python
# Question 3
d = {(1, 2): "tuple", "str": "string"}
print("Dictionary:", d)
print()

# Tuples are hashable (immutable)
print("Adding tuple key (hashable): Success")
print()

# Lists are NOT hashable (mutable)
try:
    d[[1, 2]] = "list"
except TypeError as e:
    print(f"Error with list key: {e}")
    print()

print("EXPLANATION:")
print("\u2713 Hashable (can be keys): int, str, tuple, frozenset")
print("\u2717 Not hashable (can't be keys): list, dict, set")
print("- Reason: Keys must be immutable (unchangeable)")
```

## Question 4: String Immutability

**Problem:** Strings cannot be modified in place

```python
# Question 4
s = "hello"

try:
    s[0] = "H"
except TypeError as e:
    print(f"Error: {e}")
    print()

print("EXPLANATION:")
print("Strings are IMMUTABLE - cannot change individual characters")
print()
print("Workarounds:")
s = "hello"
s = s[0].upper() + s[1:]  # Slice and concatenate
print(f"Method 1 (slice): {s}")

s = "hello"
s = "H" + s[1:]
print(f"Method 2 (concatenate): {s}")

s = "hello"
s_list = list(s)  # Convert to list
s_list[0] = "H"
s = "".join(s_list)
print(f"Method 3 (list conversion): {s}")
```

## Question 5: Mutable Default Arguments (DANGEROUS!)

**Problem:** Default mutable arguments persist between calls

```python
# Question 5 - DANGEROUS
def add_to_list_bad(item, my_list=[]):
    my_list.append(item)
    return my_list

print("Call 1:", add_to_list_bad(1))
print("Call 2:", add_to_list_bad(2))
print("Call 3:", add_to_list_bad(3))
print()
print("EXPLANATION:")
print("Default arguments are created ONCE when function is defined")
print("The same list object is reused across all calls!")
print()
print("="*50)
print("CORRECT APPROACH:")
print("="*50)

def add_to_list_good(item, my_list=None):
    if my_list is None:
        my_list = []
    my_list.append(item)
    return my_list

print("Call 1:", add_to_list_good(1))
print("Call 2:", add_to_list_good(2))
print("Call 3:", add_to_list_good(3))
```

## Question 6: List Slicing Edge Cases

**Problem:** Understanding advanced slicing

```python
# Question 6
lst = [0, 1, 2, 3, 4, 5]

print(f"Original list: {lst}")
print()
print(f"lst[2:4]: {lst[2:4]}")
print("  - Start at index 2, stop BEFORE index 4")
print()
print(f"lst[::-1]: {lst[::-1]}")
print("  - Step of -1 means reverse")
print("  - Format: [start:stop:step]")
print()
print(f"lst[10:]: {lst[10:]}")
print("  - Start at index 10 (doesn't exist): empty list")
print("  - Slicing doesn't raise error for out-of-range")
print()
print("OTHER SLICING EXAMPLES:")
print(f"lst[::2]: {lst[::2]}  - Every 2nd element")
print(f"lst[1::2]: {lst[1::2]}  - Every 2nd element starting at index 1")
print(f"lst[:3]: {lst[:3]}  - First 3 elements")
print(f"lst[-2:]: {lst[-2:]}  - Last 2 elements")
```

## Question 7: Nested List Modification

**Problem:** References in nested structures

```python
# Question 7
matrix = [[1, 2], [3, 4]]
print(f"Original matrix: {matrix}")
print()

row = matrix[0]  # row references the first inner list
print(f"row (reference to matrix[0]): {row}")
print()

row[0] = 99  # Modifying through row
print(f"After row[0] = 99:")
print(f"row: {row}")
print(f"matrix: {matrix}")
print()
print("EXPLANATION:")
print(f"row is matrix[0]: {row is matrix[0]}")
print("- row is a reference to the same object")
print("- Modifying row modifies the matrix")
```

## Question 8: Loop Variable Scope

**Problem:** Python's loop variables don't have their own scope

```python
# Question 8
for i in range(5):
    pass

print(f"Value of i after loop: {i}")
print()
print("EXPLANATION:")
print("In Python, loop variables don't create their own scope")
print("Unlike some languages (Java, C++), i remains in scope after the loop")
print(f"Final value is: {i}")
print()
print("COMPARISON:")
j = 100
for j in range(5):
    pass
print(f"j was 100, after loop: {j}")
print("The loop variable overwrites any existing variable with same name")
```

## Question 9: Truthiness & Empty Collections

**Problem:** Understanding truthy/falsy values

```python
# Question 9
print("Truthiness values:")
print(f"bool([]): {bool([])}  - Empty list is False")
print(f"bool([1]): {bool([1])}  - Non-empty list is True")
print()
print(f"bool(''): {bool('')}  - Empty string is False")
print(f"bool('0'): {bool('0')}  - Non-empty string is True")
print()
print(f"bool(0): {bool(0)}  - Zero is False")
print(f"bool(1): {bool(1)}  - Non-zero is True")
print()
print(f"bool(None): {bool(None)}  - None is False")
print(f"bool(False): {bool(False)}  - False is False")
print()
print("In if statements:")
if []:
    print("Empty list is True")
else:
    print("Empty list is False")

if [1, 2]:
    print("Non-empty list is True")

print()
print("FALSY VALUES: 0, '', [], {}, None, False, set()")
print("TRUTHY VALUES: Everything else")
```

## Question 10: String Operations & F-strings

**Problem:** Different ways to format strings

```python
# Question 10
name = "Alice"
age = 30
pi = 3.14159265

print("Method 1: % formatting (old style)")
print("Name: %s, Age: %d" % (name, age))
print()

print("Method 2: .format()")
print("Name: {}, Age: {}".format(name, age))
print("Name: {n}, Age: {a}".format(n=name, a=age))
print()

print("Method 3: f-strings (modern, Python 3.6+)")
print(f"Name: {name}, Age: {age}")
print()

print("F-string number formatting:")
print(f"Pi rounded to 2 decimals: {pi:.2f}")
print(f"Age padded to 5 digits: {age:05d}")
print(f"Pi in scientific notation: {pi:.2e}")
print()

print("F-string expressions inline:")
print(f"Age next year: {age + 1}")
print(f"Name upper: {name.upper()}")
print()

print("EXPLANATION:")
print("- f-strings are fastest and most readable (preferred in modern Python)")
print("- % formatting is legacy Python 2 style, still works but discouraged")
print("- .format() is flexible but more verbose than f-strings")
print("- Format specs like :.2f, :05d work the same across f-strings and .format()")
```
